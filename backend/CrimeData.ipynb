```python
import pandas as pd
import numpy as np
```

```python
cols=['latitude','longitude','date','time','crimetype']
```

```python
lat=np.random.uniform(low=18.4151,high=18.6352,size=10000);lat
```

```python
long=np.random.uniform(low=73.0447,high=73.9870,size=10000);long
```

```python
crime=np.random.randint(1,6,10000)
```

```python
count=np.ones(10000)
```

```python
def random_date(start, end, position=None):
    start=start.value//10**9
    end=end.value//10**9
    return pd.to_datetime(np.random.randint(start,end,10000),unit='s')
start=pd.to_datetime("2017-01-01")
end=pd.to_datetime("2018-01-01")
ls=random_date(start,end)
```

```python
df=pd.DataFrame({
    'latitude':lat,
    'longitude':long,
    'data':ls,
    'crime':crime,
    'count':count
})
#1:rape,2:murder,3:dacoity,4:kidnap,5:dowry
df.head()
```

```python
df['crime'].unique()
```

```python
df['date']=df['data'].dt.date
```

```python
df['time']=df['data'].dt.hour
```

```python
df.head()
```

```python
df.drop('data',axis=1,inplace=True)
df.drop('date',axis=1,inplace=True)
```

```python
df['latitude']=np.floor_divide(df.latitude,0.01)/100
```

```python
df['longitude']=np.floor_divide(df.longitude,0.01)/100
```

```python
df.T.to_dict()
```
